```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

###### Import csv and convert Date column to datetime, and set as index.

```python
df = pd.read_csv('prices.csv', parse_dates=['Date'], index_col=['Date'])
```

###### Filter to price and quantity of similar items, and separate periods of time.

```python
df = df.filter(items=['Regulus Price', 'Regulus Quantity Listed', 'Bast Price', 'Bast Quantity Listed', 'Wood Price', 'Wood Quantity Listed', 'Ochre Price', 'Ochre Quantity Listed', 'Resin Price', 'Resin Quantity Listed', 'Rune Price', 'Rune Quantity Listed'])
df = df.dropna()
df_2020 = df.loc['2020':'2021']
df_2022 = df.loc['2022':]
```

```python
plt.plot(df_2020['Regulus Price'], label='Regulus', linewidth=1, marker='.', color='blue')
plt.plot(df_2020['Bast Price'], label='Bast', linewidth=1, marker='.', color='magenta')
plt.plot(df_2020['Ochre Price'], label='Ochre', linewidth=1, marker='.', color='goldenrod')
plt.plot(df_2020['Resin Price'], label='Resin', linewidth=1, marker='.', color='yellowgreen')
plt.ylabel('Price (gold)')
plt.xlabel('Date')
plt.title('Regulus, Bast, Ochre, & Resin Prices 2020-2021')
plt.legend(loc="upper left")
plt.ylim([0, 120])
plt.gcf().set_size_inches(12, 8)
plt.show()
```

```python
plt.plot(df_2020['Wood Price'], label='Wood', linewidth=1, marker='.', color='sienna')
plt.plot(df_2020['Rune Price'], label='Rune', linewidth=1, marker='.', color='tomato')
plt.ylabel('Price (gold)')
plt.xlabel('Date')
plt.title('Wood & Rune Prices 2020-2021')
plt.legend(loc="upper left")
plt.ylim([0, 1200])
plt.gcf().set_size_inches(12, 8)
plt.show()
```
