```python
# import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from matplotlib.patches import Rectangle
import matplotlib.dates as mdates

plt.style.use('ggplot')
```

```python
# import csv and convert 'Date' column to datetime, and set as index
df = pd.read_csv('prices.csv', parse_dates=['Date'], index_col=['Date'])
```

```python
# filter to price & quantity of similar items.
price = df.filter(items=['Date', 'Regulus Price', 'Bast Price', 'Wood Price', 'Resin Price', 'Rune Price', 'Wax Price'])
quantity = df.filter(items=['Date', 'Regulus Quantity Listed', 'Bast Quantity Listed', 'Wood Quantity Listed', 'Resin Quantity Listed', 'Rune Quantity Listed', 'Wax Quantity Listed'])

# simplify column names
cols = ['Regulus', 'Bast', 'Wood', 'Resin', 'Rune', 'Wax']
price.columns = cols
quantity.columns = cols

# create separate dfs for continous periods of time
price_2020 = price.loc['2020':'2021']
quantity_2020 = quantity.loc['2020':'2021']
price_2022 = price.loc['2022':]
quantity_2020 = quantity.loc['2022':]

# remove missing values
price = price.dropna()
quantity = quantity.dropna()
```

```python
# find min and max prices for 2020 and 2022
mat_idxmin20 = price_2020.idxmin()
mat_min20 = price_2020.min()
mat_idxmin22 = price_2022.idxmin()
mat_min22 = price_2022.min()
mat_idxmax20 = price_2020.idxmax()
mat_max20 = price_2020.max()
mat_idxmax22 = price_2022.idxmax()
mat_max22 = price_2022.max()

# combine into dataframe
mat_info = pd.concat([mat_idxmin20, mat_min20, mat_idxmax20, mat_max20, mat_idxmin22, mat_min22, mat_idxmax22, mat_max22], axis=1)

# label columns
cols = ['Min Date A', 'Min A', 'Max Date A', 'Max A', 'Min Date B', 'Min B', 'Max Date B', 'Max B']
mat_info.columns = cols

display(mat_info)
```

```python
# find amount of change between entries
price_diff = price.diff()

# remove start + first date after large gaps
price_diff = price_diff.dropna()
price_diff = price_diff.drop(labels=['2020-09-21', '2022-03-19'])


# descriptive statistics for price changes
price_diff_desc = price_diff.describe()
price_diff_desc = price_diff_desc.round(decimals = 0)


display(price_diff_desc)
```

```python
quant_diff = quantity.diff()

quant_diff = quant_diff.dropna()
quant_diff = quant_diff.drop(labels=['2022-03-19'])

quant_diff_desc = quant_diff.describe()
quant_diff_desc = quant_diff_desc.round(decimals = 0)

display(quant_diff_desc)
```

```python
# dictionaries for plot colors, fonts etc
ffont = {'fontsize':6, 'color':'#ff6d00'}
ffont2 = {'fontsize':6, 'color':'#58BABB'}
xfont = {'fontsize':5, 'color':'silver'}
ftitle = {'fontsize':7, 'color':'slategray'}

focus = {'lw':1.5, 'color':'#ff6d00', 'alpha':0.95}
focus2 = {'lw':2, 'color':'#58BABB', 'alpha':0.55}
neutral = {'lw':0.5, 'color':'slategray', 'alpha':0.5}

event = {'ec':None, 'color':'w', 'alpha':0.5}
expl = {'ec':None, 'color':'honeydew', 'alpha':0.5}
update = {'lw':1, 'linestyle':'dotted', 'c':'gray', 'alpha':0.6}

fgrid = {'grid_color':'#FFF2E8', 'grid_alpha':0.5, 'labelsize':3.5, 'size':0}
fgrid2 = {'grid_color':'#EDFFFF', 'grid_alpha':0.5, 'labelsize':3.5, 'size':0}
xticks = {'color':'w', 'direction':'in', 'labelsize':3.5, 'size':2}
```

```python
# create a row of 3 subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, gridspec_kw={'width_ratios':[1.39,5,4.8]})
fig.set_size_inches(5,2.5)
fig.set_dpi(200)
fig.suptitle('elder scrolls online crafting material prices', **ftitle)

# plot prices
ax1.set_title('2019', pad=-2, **xfont)
ax1.plot(price['Wood'], **focus)
ax1.plot(price['Regulus'], **neutral)
ax1.plot(price['Bast'], **neutral)
ax1.plot(price['Resin'], **neutral)
ax1.plot(price['Rune'], **neutral)
ax1.plot(price['Wax'], **neutral)

ax2.set_title('2020 - 2021', pad=-2, **xfont)
ax2.plot(price['Wood'], **focus)
ax2.plot(price['Regulus'], **neutral)
ax2.plot(price['Bast'], **neutral)
ax2.plot(price['Resin'], **neutral)
ax2.plot(price['Rune'], **neutral)
ax2.plot(price['Wax'], **neutral)

ax3.set_title('2022', pad=-2, **xfont)
ax3.plot(price['Wood'], label='wood', **focus)
ax3.plot(price['Regulus'], label='regulus', **neutral)
ax3.plot(price['Bast'], label='bast', **neutral)
ax3.plot(price['Resin'], label='resin', **neutral)
ax3.plot(price['Rune'], label='rune', **neutral)
ax3.plot(price['Wax'], label='wax', **neutral)
ax3.legend(facecolor='w', edgecolor='w', loc=(0.72,0.135), fontsize=4)

# set x-axis limits for each plot
ax1.set_xlim(datetime(2019,11,4,0,0,0), datetime(2019,12,12,0,0,0))
ax2.set_xlim(datetime(2020,9,20,0,0,0), datetime(2021,2,5,0,0,0))
ax3.set_xlim(datetime(2022,3,18,0,0,0), datetime(2022,7,19,0,0,0))

# lines/rectangles for holiday event & patches
# update 28
ax2.axvline(x=['2020-11-02'], **update)

# new life 2020
xy = datetime(2020,12,17,0,0,0), ax2.get_ylim()[0]
w, h = timedelta(days=19), ax2.get_ylim()[1] - ax2.get_ylim()[0]
ax2.add_patch(Rectangle(xy, w, h, **event))

# update 34
ax3.axvline(x=['2022-06-06'], **update)

# hide inner spines
ax1.spines['right'].set_visible(False)
ax2.spines[['left', 'right']].set_visible(False)
ax3.spines['left'].set_visible(False)

# remove inner axis ticks/labels
ax1.yaxis.tick_left()
ax1.tick_params(labelsize=3.5, size=0)
ax2.tick_params(labelsize=3.5, size=0)
ax3.yaxis.tick_right()
ax3.tick_params(labelright='off', labelsize=3.5, size=0)

# adjust date tick labels
ax1.xaxis.set_major_locator(mdates.DayLocator(interval=14))
ax2.xaxis.set_major_locator(mdates.DayLocator(interval=14))
ax3.xaxis.set_major_locator(mdates.DayLocator(interval=14))
ax1.xaxis.set_major_formatter(mdates.DateFormatter("%m/%d"))
ax2.xaxis.set_major_formatter(mdates.DateFormatter("%m/%d"))
ax3.xaxis.set_major_formatter(mdates.DateFormatter("%m/%d"))

# squish plots closer together
fig.subplots_adjust(wspace=0.04)

plt.show()
```

```python
# create a row of 3 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, gridspec_kw={'width_ratios':[5,4.8]})
fig.set_size_inches(5,2.5)
fig.set_dpi(200)
fig.suptitle('eso crafting material: total quantity listed', **ftitle)

# quantity plots
ax1.set_title('2020 - 2021', pad=-2, **xfont)
ax1.plot(quantity['Wood'], **focus)
ax1.plot(quantity['Regulus'], **neutral)
ax1.plot(quantity['Bast'], **neutral)
ax1.plot(quantity['Resin'], **neutral)
ax1.plot(quantity['Rune'], **neutral)
ax1.plot(quantity['Wax'], **neutral)

ax2.set_title('2022', pad=-2, **xfont)
ax2.plot(quantity['Wood'], label='wood', **focus)
ax2.plot(quantity['Regulus'], label='regulus', **neutral)
ax2.plot(quantity['Bast'], label='bast', **neutral)
ax2.plot(quantity['Resin'], label='resin', **neutral)
ax2.plot(quantity['Rune'], label='rune', **neutral)
ax2.plot(quantity['Wax'], label='wax', **neutral)
ax2.legend(facecolor='w', edgecolor='w', loc='upper right', fontsize=4)

# set x-axis limits for each plot
ax1.set_xlim(datetime(2020,9,20,0,0,0), datetime(2021,2,5,0,0,0))
ax2.set_xlim(datetime(2022,3,18,0,0,0), datetime(2022,7,17,0,0,0))

# lines/rectangles for holiday event & patches
# update 28
ax1.axvline(x=['2020-11-02'], **update)

# explorer's celebration 2020
xy = datetime(2020,11,12,0,0,0), ax1.get_ylim()[0]
w, h = timedelta(days=5), ax1.get_ylim()[1] - ax1.get_ylim()[0]
ax1.add_patch(Rectangle(xy, w, h, **expl))

# new life 2020
xy = datetime(2020,12,17,0,0,0), ax1.get_ylim()[0]
w, h = timedelta(days=19), ax1.get_ylim()[1] - ax1.get_ylim()[0]
ax1.add_patch(Rectangle(xy, w, h, **event))

# update 34
ax2.axvline(x=['2022-06-06'], **update)

# hide inner spines
ax1.spines['right'].set_visible(False)
ax2.spines[['left', 'right']].set_visible(False)

# remove inner axis ticks/labels
ax1.yaxis.tick_left()
ax1.tick_params(labelsize=3.5, size=0)
ax2.yaxis.tick_right()
ax2.tick_params(labelright='off', labelsize=3.5, size=0)

# adjust date tick labels
ax1.xaxis.set_major_locator(mdates.DayLocator(interval=14))
ax2.xaxis.set_major_locator(mdates.DayLocator(interval=14))
ax1.xaxis.set_major_formatter(mdates.DateFormatter("%m/%d"))
ax2.xaxis.set_major_formatter(mdates.DateFormatter("%m/%d"))

# squish plots closer together
fig.subplots_adjust(wspace=0.02)

plt.show()
```

```python
# Line graph showing relationship between changes in wood price and quantity listed for sale

fig, ax1 = plt.subplots()
fig.suptitle('change in price vs quantity: wood', **ftitle)
fig.set_size_inches(3, 1.5)
fig.set_dpi(200)

ax1.plot(price_diff['Wood'], **focus)
ax1.set_ylabel('price change', **ffont)
ax1.set_xlabel('2020 - 2021', **xfont)
ax1.tick_params(axis='y', **fgrid)
ax1.tick_params(axis='x', **xticks)
ax1.xaxis.set_major_formatter(mdates.DateFormatter("%m-%d"))
ax1.set_xlim(datetime(2020,9,24,0,0,0), datetime(2021,2,5,0,0,0))
ax1.set_ylim(bottom=-185)

ax2 = ax1.twinx()
ax2.plot(quant_diff['Wood'], **focus2)
ax2.set_ylabel('quantity change', **ffont2)
ax2.tick_params(axis='y', **fgrid2)
ax2.xaxis.set_major_formatter(mdates.DateFormatter("%m/%d"))

plt.show()
```
